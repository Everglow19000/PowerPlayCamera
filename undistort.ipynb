```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
# Camera calibration
f1, f2 = 1385.92, 1385.92
Cx, Cy = 951.982, 534.084
mtx = np.array([[f1, 0, Cx],
                [0, f2, Cy],
                [0, 0, 1]])
dist = np.array(
    [[0.117627, -0.248549, 0, 0, 0.107441, 0, 0, 0]])
```

```python
img = cv2.imread("src/9.jpeg")
```

```python
img_undistorted = cv2.undistort(img, mtx, dist, None, mtx)
```

```python
img_rgb = cv2.cvtColor(img_undistorted, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```
