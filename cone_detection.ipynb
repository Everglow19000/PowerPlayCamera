```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import colorsys
```

```python
img = cv2.imread('./src/8.jpeg')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
msk = cv2.inRange(img_hsv, np.array([-20, 80, 50]), np.array([10, 255, 255]))
krn = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 3))
dlt = cv2.dilate(msk, krn, iterations=2)

img_edges = cv2.Canny(dlt, 80, 160)

contours, hierarchy = cv2.findContours(img_edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img_contours = np.zeros_like(img)
print(len(contours))

for i in range(len(contours)):
	(h, s, v) = ((i/30) % 1, 1, 0.9)
	(r, g, b) = colorsys.hsv_to_rgb(h, s, v)
	cv2.drawContours(img_contours, contours, i, (r * 255, g * 255, b * 255), 3)

plt.imshow(img_contours)
```
